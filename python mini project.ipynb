```python
x = pd.read_csv('messy_sales_dataset.csv')
x
```

```python
x.isnull().sum()
```

```python
x['TotalSales'] = pd.to_numeric(x['TotalSales'],errors='coerce')
```

```python
x['UnitPrice'] = pd.to_numeric(x['UnitPrice'],errors='coerce')
```

```python
x['Quantity'] = pd.to_numeric(x['Quantity'],errors='coerce')
```

```python
x[x.isnull().any(axis=1)]
```

```python
x['TotalSales'] = x['TotalSales'].fillna(x['TotalSales'].mean())
```

```python
x['Quantity'] = x['Quantity'].fillna(x['Quantity'].median())
```

```python
x['UnitPrice'] = x['UnitPrice'].fillna(x['UnitPrice'].mean())
```

```python
xdr = x.dropna(subset='Region')
```

```python
xdp = x.dropna(subset='Product')
```

```python
x[x.isnull().any(axis=1)]
```

```python
x['Region'] = x['Region'].fillna('Unknown')
x['Product'] = x['Product'].fillna('Unknown')
```

```python
x[x.isnull().any(axis=1)]
```

```python
x['TotalSales'] = x['Quantity']*x['UnitPrice']
```

```python
x.duplicated().sum()
```

```python
x=x.drop_duplicates()
```

```python
x.duplicated().sum()
```

```python
x['Date'] = pd.to_datetime(x['Date'], errors='coerce')
```

```python
x['Month'] = x['Date'].dt.month_name()
x['Year'] = x['Date'].dt.year
```

```python
x= x.dropna(subset=['Date'])
```

```python
# EDA Analysis
```

```python
x = pd.read_csv('cleaned Sales DAtaset.csv')
x
```

```python
# total Sales Oveerview

totalsales = x['TotalSales'].sum()
totalorders = x['OrderID'].nunique()

print("Total Sales:", totalsales)
print("Total Orders:", totalorders)
```

```python
#sales by product

salesbyproduct = x.groupby('Product')['TotalSales'].sum().sort_values(ascending=False)
print(salesbyproduct)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
salesbyproduct.plot(kind='bar')
```

```python
# sales by region

sr = x.groupby('Region')['TotalSales'].sum()
print(sr)
```

```python
sr.plot(kind='bar')
```

```python
# monthly sales trend

ms = x.groupby('Month')['TotalSales'].sum()
print(ms)
```

```python
# Top 5 orders by Revenue

top5 = x.sort_values(by='TotalSales',ascending=False).head(5)
print(top5)
```

```python
# Average Product Price

app = x.groupby('Product')['UnitPrice'].mean()
print(app)
```

```python
# product demand (Quantity Sold) 
pd = x.groupby('Product')['Quantity'].sum()
print(pd)
```

```python
# Sales table

pivot = x.pivot_table(values='TotalSales',index='Region',columns='Product',aggfunc='sum')
print(pivot)
```
